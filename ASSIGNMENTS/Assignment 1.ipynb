```python
ENG = int(input("Enter marks of subject ENG: "))
URDU = int(input("Enter marks of subject URDU: "))
MATH = int(input("Enter marks of subject MATH: "))
SCI = int(input("Enter marks of subject SCI: "))
SINDHI = int(input("Enter marks of subject SINDHI: "))

total = ENG + URDU + MATH + SCI + SINDHI
percentage = total / 5

print("Total =", total)
print("Percentage =", percentage)

if percentage >= 80:
    print("Grade = A")
elif percentage >= 70:
    print("Grade = B")
elif percentage >= 60:
    print("Grade = C")
elif percentage >= 50:
    print("Grade = D")
else:
    print("Grade = Fail")
```
